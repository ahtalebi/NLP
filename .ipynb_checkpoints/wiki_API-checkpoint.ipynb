```python
import spacy       # is for NLP
nlp = spacy.load('en_core_web_md')   #   en_core_web_md is a pre-trained spaCy model for English language text processing

sen = nlp('the 2004 united states presidential election was the 55th quadrennial presidential election')
sen
```

```python
from spacy.lang.en.stop_words import STOP_WORDS as stopwords
import string

punctu = string.punctuation
```

```python
def text_cleaner(sen):
    cleaned_sen = sen.replace("'s", "")
    return cleaned_sen
```

```python
L1 = text_cleaner("The 2020 United States elections were held on! Tuesday, November 3, 2020. The Democratic Party\'s")
L1
```

## Tokenize
## Lemmatise (into root form) and lowercase
## Removing stopwords and punctuations

```python
def tokenizer(text):
    w1 = nlp(text)
    w2 = [w.lemma_.lower().strip() if w.lemma_ != "-PRON-" else w.lower_ for w in w1]
    w3 = [w for w in w2 if (w not in stopwords and w not in punctu)]
    return w3

tokenizer(L1)
```

## syntax for list comprehension

new_list = [expression for item in iterable if condition]


## now we can combine, text_cleaner and tokenizer functions

```python
# this func is better for one sentence. 

def tokenize_clean(sen):
    text = sen.replace("'s", "")
    w1 = nlp(text)
    w2 = [w.lemma_.lower().strip() if w.lemma_ != "-PRON-" else w.lower_ for w in w1]
    w3 = [w for w in w2 if (w not in stopwords and w not in punctu)]
    return w3

L2 = tokenize_clean("The 2020 United States elections were held on! Tuesday, November 3, 2020. The Democratic Party\'s")
L2
```

```python
cleaned_1 = ' '.join([text_cleaner(s) for s in myfile.values()])  # we use ''.join([]) to convert list to string
type(cleaned_1)
```

```python
cleaned_2 = tokenizer(cleaned_1)
```

```python
cleaned_2
```

```python
## function which does the cleaning and tokenization at the same time is here :

def file_clean_toekenize (f_text_dic):
    cleaned_1 = ' '.join([text_cleaner(s) for s in f_text_dic.values()])
    cleaned_2 = tokenizer(cleaned_1)
    return cleaned_2
```

```python
file_clean_toekenize(myfile)
```

```python
def word_map(f_text_dic):
    word_counts = {}
    word_list = file_clean_toekenize(f_text_dic)

    for w in word_list:
        word_counts[w] =  word_counts.get(w, 0) + 1
        sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)   # Sort the word counts in descending order
    return word_counts

word_map(myfile)
```

```python
def word_map(f_text_dic):
    word_counts = {}
    word_list = file_clean_toekenize(f_text_dic)

    for w in word_list:
        word_counts[w] = word_counts.get(w, 0) + 1

    sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    return sorted_word_counts
```

```python
ordered_word_counts = word_map(myfile)

for word, count in ordered_word_counts:
    print(f"{word}: {count}")
```

```python
ordered_word_counts

word_counts_dict = {word: count for word, count in ordered_word_counts}
word_counts_dict
```

```python
import matplotlib.pyplot as plt

sorted_word_frequencies = sorted(word_counts_dict.values(), reverse=True)

plt.figure(figsize=(10, 6))
plt.loglog(range(1, len(sorted_word_frequencies) + 1), sorted_word_frequencies, marker='o', linestyle='None', markersize=5)
plt.title('Log-log plot of word frequencies')
plt.xlabel('Rank of word (log scale)')
plt.ylabel('Word frequency (log scale)')
#plt.grid(True, which="both", ls="--", color='gray', alpha=0.5)
plt.show()
```

## X-Axis (Rank of Word): The x-axis represents the rank of each word in the corpus. The word with the highest frequency is assigned a rank of 1, the second highest frequency is assigned a rank of 2, and so on. 
## Y-Axis (Word Frequency): The y-axis represents the frequency of each word in the corpus. It shows how many times each word appears in the text.

## Zipf`s Law:

$ f(r) = \frac{k}{r^s} $

Where:
- \( f(r) \) is the frequency of the word at rank \( r \).
- \( k \) is a constant.
- \( s \) is the scaling parameter, typically close to 1.
